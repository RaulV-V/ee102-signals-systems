```python
import numpy as np
import matplotlib.pyplot as plt
# set math fonts in matplotlib to true
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'serif'

# Define x(t) = sin(t)
def x(t):
    return np.sin(t)

# Define your time axies
time_data = np.linspace(-10, 10, 1000)
x_original = x(time_data)
# Compute various signal transformations
x_reversed = x(-time_data) # time reversal
# the "3" is not time..we need shift in time units, not indices
x_shifted_right = x(time_data - np.pi/2) # delay
x_shifted_left = x(time_data + np.pi/2) # advance
x_scaled_compressed = x(2 * time_data) # scale (compress)
x_scaled_expanded = x(0.5 * time_data) # scale (expand)
x_combined_1 = x(2 * time_data - 3) # scale first (compress), then shift
x_combined_2 = x(2 * (time_data + 3)) # shift first, then scale
```

```python
from utils import plot_signal_transformations
plots = [
        (0, 0, x_original, '$x(t) = \sin(t)$: Original'),
        (0, 1, x_reversed, '$\\bar{x}(t) = x(-t)$: Time Reversal'),
        (0, 2, x_shifted_right, '$\\bar{x}(t) = x(t-\pi/2)$: Delay'),
        (0, 3, x_shifted_left, '$\\bar{x}(t) = x(t+\pi/2)$: Advance'),
        (1, 0, x_scaled_compressed, '$\\bar{x}(t) = x(2t)$: Compressed'),
        (1, 1, x_scaled_expanded, '$\\bar{x}(t) = x(0.5t)$: Expanded'),
        (1, 2, x_combined_1, '$\\bar{x}(t) = x(2t-\pi/2)$: Combined'),
        (1, 3, x_combined_2, '$\\bar{x}(t) = x(2(t+\pi/2))$: Combined')
]
plot_signal_transformations(time_data, plots)
```
